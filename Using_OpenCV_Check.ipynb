```python
import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Let's do some basic operations using OpenCV
# Reading an image
img = cv2.imread('brown_bear.jpeg')# You can replace this file path by your own file path.
print (img.shape)
```

### cv2.imshow() is also there but works well only with terminal, does not work that well with jupyter notebook, hence using plt.imshow()

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # converting BGR to RGB for using matplotlib
plt.show()
```

```python
# Now, let's say we want to demonstrate only the blue channel out of these
# The ordering of channels in OpenCV is B-G-R. 0 is Blue, 1 is Green, 2 is Red.
plt.imshow(img[:,:,0])
plt.show()
```

```python
# Now, let's say we want to demonstrate only the red channel out of these
# The ordering of channels in OpenCV is B-G-R. 0 is Blue, 1 is Green, 2 is Red.
plt.imshow(img[:,:,2])
plt.show()
```

```python
# Now, let's say we want to demonstrate only the green channel out of these
# The ordering of channels in OpenCV is B-G-R. 0 is Blue, 1 is Green, 2 is Red.
plt.imshow(img[:,:,1])
plt.show()
```

```python
# This image is a colour image. We can easily convert this image to gray.
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
# Next, we can access specific parts of the image using coordinates
bearface = img[100:450, 150:550]
plt.imshow(bearface)
plt.show()
```
